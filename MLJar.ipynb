# Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from supervised.automl import AutoML

from sklearn.model_selection import train_test_split
```

# Data

```python
df = pd.read_csv('https://query.data.world/s/5cqzv6uhbwh6bjphv3kea6yo2qbr4t', engine='python', encoding='latin-1')
```

```python
df
```

```python
df['binnedInc'] = df.binnedInc.astype('category')
df['Geography'] = df.Geography.str.split(',', expand=True)[1]
```

# Train - Test

```python
X = df.copy().drop(columns=['TARGET_deathRate'])
y = df['TARGET_deathRate']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# Explain

```python
automl_e = AutoML(mode="Explain")
automl_e.fit(X_train, y_train)
```

```python
!zip -r automl_e.zip AutoML_1 
```

# Compete

```python
automl_c = AutoML(mode="Compete")
automl_c.fit(X_train, y_train)
```
